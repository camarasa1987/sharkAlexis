```python
def age_of_victims(age): # funcion para agrupar y asi reducir la cantidad de edades distintas en grupos por edad

        if age == 0:

            return 'No information'

        if age <= 12:

            return 'kid'

        elif age <= 19:

            return 'teen'

        elif age <= 60:

            return 'adult'

        elif age >= 61:

            return 'elderly'

        else:
            return 'No information'

ataque['Unnamed_22'] = ataque['Age'].apply(age_of_victims)
```

```python
def time_of_day(hour): # funcion para ag
    if '07' in hour or '08' in hour or '09rupar y asi reducir la cantidad de horas del dia en periodos preestablecidos
' in hour or '10' in hour or '11' in hour or '12' in hour:
        return 'Morning'
    elif '13' in hour or '14' in hour or '15' in hour or '16' in hour or '17' in hour or '18' in hour or '19' in hour or '20' in hour or '21' in hour:
        return 'Afternoon'
    elif '22' in hour or '23' in hour or '00' in hour or '01' in hour or '02' in hour or '03' in hour or '04' in hour or '05' in hour or '06' in hour:
        return 'Night'
    else:
        return 'No information'

ataque['Unnamed_23'] = ataque['Time'].apply(time_of_day)
```

```python
def year_to_century(year): # funcion para agrupar por siglos

        if year <= 1499:

            return 'No Concrete information'

        elif year <= 1599:

            return 'XVI'

        elif year <= 1699:

            return 'XVII'
        
        elif year <= 1799:

            return 'XVIII'
        
        elif year <= 1899:

            return 'XIX'
        
        elif year <= 1999:

            return 'XX'
        
        elif year <= 2099:

            return 'XXI'

        else:
            return 'No information'

ataque['Case_Number1'] = ataque['Year'].apply(year_to_century)
```
